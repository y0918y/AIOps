```python
cur_dir = os.getcwd()
files = os.path.join(cur_dir, '*.txt')
file_list = glob.glob(files)

file_list
```

```python
# 국립국어원 언어정보 말뭉치 가져오기

import codecs
from bs4 import BeautifulSoup
from konlpy.tag import Okt

# utf-16 인코딩으로 파일을 열고 글자를 출력하기 

fp = codecs.open(file_list[0], 'r', encoding='utf-16')
soup = BeautifulSoup(fp, 'html.parser')
body = soup.select_one('body > text')
text = body.getText()
```

```python
text
```

```python
text.split('wn')[0].split('\n')[1]
```

```python
# 텍스트를 한줄씩 처리하기

okt = Okt()
word_dic = { }

lines = text.split('\n')

for line in lines :
    malist = okt.pos(line)
    for word in malist :
        if word[1] == 'Noun' : # 명사 확인하기
            if not (word[0] in word_dic) :
                word_dic[word[0]] = 0
            else : 
                word_dic[word[0]] += 1 # 카운트 1하기
```

```python
word_dic.items()
```

```python
# 많이 사용된 명사 출력하기

keys = sorted(word_dic.items(), key = lambda x : x[1], reverse = True) # sorted 함수 -> (iterable, 정렬의 key, 내림차순 여부)

for word, count in keys[ : 50 ] :
    print('{0} : ({1})'.format(word, count), end='')
```

```python
# 텍스트 정제 함수 : 한글 이외의 문자는 전부 제거한다

def text_cleaning(text) :

    import re
    
    hangul = re.compile('[가-힣]+')
    result = hangul.findall(text)
    return result
```

```python
결과파일 = text_cleaning(text.split('wn')[0])
```

```python
결과파일[0:5]
```

```python
from collections import Counter

# 추출한 토큰이 각각 몇개 등장했는지 개수를 세어준다
count = Counter(결과파일)
count
```

```python
import pandas as pd

결과파일_df = pd.DataFrame( { 'word' : count.keys(), 
                         'count' : count.values() } )
결과파일_df.head()
```

```python
결과파일_df.sort_values(by = 'count', ascending=False)
```

```python
# stopwords 제거하기

# 불필요한 단어 리스트 만들기

stopwords = [ '그', '안', '것' ]

# 결과파일에서 stopwords 리스트에 해당하는 단어들 제거하기
# Counter 함수 이용해서 단어와 갯수로 dict를 만들고 dict comprehension의 key-value 속성을 이용해 컨트롤하기
# 이때 이미 만들어 놓은 count를 활용한다!

remove_결과파일_count = Counter( { x : count[x] for x in count if x not in stopwords })

remove_결과파일_df = pd.DataFrame( { 'word' : remove_결과파일_count.keys(), 
                         'count' : remove_결과파일_count.values() } )
remove_결과파일_df.sort_values(by='count', ascending = False)
```

```python
# corpus 만들기

결과파일_corpus = ' '.join(결과파일)
결과파일_corpus
```

```python
from apyori import apriori

# 장바구니 형태의 데이터(트랜잭션 데이터)를 생성한다

transactions = [ ['복돌이', '복덩이'],
                ['복돌이', '복뎅이'],
                ['복덩이', '복동이'],
                ['복뎅이', '복동이'],
                ['복덩이', '복뎅이', '복동이'],
                ['복돌이', '복덩이', '복동이'] ]

# 연관분석을 수행한다

results = list(apriori(transactions))
for result in results :
    print(result)
```

```python
# 연관분석의 조건 설정하기

# 지지도 0.5, 신뢰도 0.6, 향상도 1.0 이상이면서
# ['복돌이', '복덩이'] 처럼 규칙의 크기가 2 이하인 규칙을 추출한다

list(apriori(transactions,
             min_support = 0.5,
             min_confidence = 0.6,
             min_lift = 1.0,
             max_length = 2))
```

```python
# 결과파일로 만든 corpus에서 명사만 추출해보자!

okt = Okt()

nouns = okt.nouns(결과파일_corpus)
nouns[ :10]
```

```python
list(apriori(nouns[ :20],
             min_support = 0.1,
             min_confidence = 0.2,
             min_lift = 1.0,
             max_length = 2))

# 아!!!! corpus가 여러개 잇으면서 -> 동시에 나오는 걸 계산해야 하니! 1개 corpus만으론 결과가 없지!!!
```
