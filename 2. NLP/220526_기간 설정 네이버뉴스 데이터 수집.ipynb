```python
import re
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm, tqdm_notebook
```

```python
key1 = '신한은행'
page = [ i for i in range(1, 4000, 10)]

Year = ['2020','2020','2021']
Month = ['01','02','03','04','05','06','07','08','09','10','11','12']
Day_S = ['1','6','11','16','21','26']
Day_E = ['5','10','15','20','25','31']

list_naver_news = [ ]
times = [ ]
titles = [ ]
links = [ ]
infos = [ ]

for year in Year:
  for month in Month:
    print(month)
    for i in range(0,1):
      Date_Start = year + '.' + month + '.' + Day_S[i]
      Date_End = year + '.' + month + '.' + Day_E[i]

      Date_Start_2 = year + month + Day_S[i]
      Date_End_2 = year + month + Day_E[i]

     # print(Date_Start, Date_End, Date_Start_2, Date_End_2)

      news_time = [ ] 
      news_title = [ ]
      news_info = [ ]

      for p in tqdm(page):
        
        url_news = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query='+key1+'&sort=2&photo=0&field=0&pd=3&ds=' + Date_Start + '&de=' + Date_End
        url_news = url_news + '&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:from' + str(Date_Start_2) + 'to' + str(Date_End_2) + ',a:all&start=' + str(p)
                
        
        r = requests.get(url_news)
        soup = BeautifulSoup(r.text, 'html.parser')

        time = soup.body.find_all('span', attrs={'class':'info'}) 
        news_time = time

        title = soup.body.find_all('a', attrs={'class':'news_tit'}) 
        news_title = title

        info = soup.body.find_all('a', attrs={'class':'api_txt_lines dsc_txt_wrap'})
        news_info = info

        if len(news_title) > 0 :
          for j in range(len(news_title)): 
              times.append(news_time[j].text)
              titles.append(news_title[j].text)
              links.append( str(news_title[j]).split('<a class="news_tit" href=')[1].split('"')[1] )
              infos.append(news_info[j].text)
        else:
          break

      print(len(titles))
#      print(titles)
#Date_Start = '2019.01.01'
#Date_End = '2019.01.15'

#Date_Start_2 = '20190101'
#Date_End_2 = '20190115'

len(list_naver_news)
```

```python
import pandas as pd

df = pd.DataFrame({ 'time' : times, 'title' : titles, 'info' : infos, 'link' : links })
df.to_excel('크롤링(신한은행,2020추가).xlsx', encoding='utf-8-sig')
```
